```python
# def get_cols_with_zero_std_dev(df:pd.DataFrame):
#     """
#     Returns a list of columns names who are having zero standard deviation.
#     """

#     cols_to_drop=[]
#     num_cols=[col for col in df.columns if df[col].dtype!='O']#numericals only
#     for col in num_cols:
#         if df[col].std()==0:
#             cols_to_drop.append(col)
#     return cols_to_drop


# import pandas as pd

# def get_redundant_cols(df: pd.DataFrame, missing_thresh: float = 0.7):
#     """
#     Return a list of columns having missing values more than a certain threshold.
#     """
#     # Calculate the ratio of missing values for each column
#     cols_missing_ratios = df.isna().sum().div(df.shape[0])
    
#     # Identify columns with missing values ratio greater than the threshold
#     cols_to_drop = cols_missing_ratios[cols_missing_ratios > missing_thresh].index.tolist()
    
#     return cols_to_drop




# ## seprate feature and labesls out


# X,y=wafers.drop(cols_to_drop,axis=1),wafers[["Good/Bad"]]




import pandas as pd

def get_cols_with_zero_std_dev(df: pd.DataFrame):
    """
    Returns a list of column names with zero standard deviation.
    """
    cols_to_drop = []
    num_cols = [col for col in df.columns if df[col].dtype != 'O']  # Numerical columns only
    for col in num_cols:
        if df[col].std() == 0:
            cols_to_drop.append(col)
    return cols_to_drop

def get_redundant_cols(df: pd.DataFrame, missing_thresh: float = 0.7):
    """
    Return a list of columns having missing values more than a certain threshold.
    """
    # Calculate the ratio of missing values for each column
    cols_missing_ratios = df.isna().sum().div(df.shape[0])
    
    # Identify columns with missing values ratio greater than the threshold
    cols_to_drop = cols_missing_ratios[cols_missing_ratios > missing_thresh].index.tolist()
    
    return cols_to_drop

# Example usage:
# Assuming `wafers` is your DataFrame
cols_zero_std_dev = get_cols_with_zero_std_dev(wafers)
cols_redundant = get_redundant_cols(wafers, missing_thresh=0.7)

# Combine lists and remove duplicates
cols_to_drop = list(set(cols_zero_std_dev + cols_redundant))

# Separate features and labels
X = wafers.drop(columns=cols_to_drop)
y = wafers[["Good/Bad"]]
```

```python
# from sklearn.pipeline import Pipeline
# from sklearn.impute import  KNNImputer
# from sklearn.impute import  SimpleImputer
# from sklearn.preprocessing import  RobustScaler
# ### from sklearn.preprocessing import StandardScaler

# imputer=KNNImputer(n_neighbors=3)
# preprocessing_pipeline=Pipeline(
#     steps=[('Imputer',imputer),('Scaler',RobustScaler())])


# preprocessing_pipeline



import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer
from sklearn.preprocessing import RobustScaler

# Define the pipeline with KNNImputer and RobustScaler
imputer = KNNImputer(n_neighbors=3)
preprocessing_pipeline = Pipeline(
    steps=[
        ('Imputer', imputer),
        ('Scaler', RobustScaler())
    ]
)

# Select only numerical columns from X
X_numeric = X.select_dtypes(include=['number'])

# Transform the numerical features
X_trans = preprocessing_pipeline.fit_transform(X_numeric)

print("Shape of transformed features set:", X_trans.shape)

# Display the transformed features
print(X_trans)
```
